```python
data_conc = pd.read_csv('tables/LSARP-SA017-SA028_concentrations.csv')
data_conc['BI_NBR'] = 'NA'
data_conc.loc[data_conc.SType == 'Bi', 'BI_NBR'] = data_conc.ms_file[data_conc.SType == 'Bi'].apply(lambda x: x[-10:])
data_conc.head()
```

```python
data_conc.sort_values(by = ['Col', 'Batch'], inplace = True)
data_conc = data_conc[data_conc.Col == 'Col02']
```

```python
data_conc.columns
```

```python
availability = data_conc.copy()

    
availability_data = pd.DataFrame()
availability_data['Compound'] = availability.columns[1:-5]
availability_data['Ave_Conc'] = 0
availability_data['Std'] = 0

for cp in availability_data.Compound:
    availability_data.loc[availability_data.Compound == cp, 'Ave_Conc'] = np.mean(availability[cp][(availability.SType == 'MHPool')])
    availability_data.loc[availability_data.Compound == cp, 'Std'] = np.std(availability[cp][availability.SType == 'MHPool'])/np.sqrt(len(availability[cp][availability.SType == 'MHPool']))

availability_data['Ave_Conc_Min'] = availability_data.Ave_Conc - availability_data.Std
availability_data['Ave_Conc_Max'] = availability_data.Ave_Conc + availability_data.Std

concentrations = availability_data.sort_values(by = 'Ave_Conc', ascending = True).reset_index(drop = True)[['Compound', 'Ave_Conc', 'Std', 'Ave_Conc_Min', 'Ave_Conc_Max']]
concentrations = concentrations[concentrations.Ave_Conc > 1]
```
